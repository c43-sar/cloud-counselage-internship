```python
import pandas as pd
import numpy as np
import math

data = pd.read_excel("Final Lead Data.xlsx")
```

```python
print(data.columns)
for i in data.columns:
    print(i)
    print(data[i].unique())
```

```python
from datetime import datetime as dt

dt_str = "%m/%d/%Y %I:%M:%S %p"
```

```python
year_arr = []
for i in range(data.shape[0]):
    year_arr.append(dt.strptime(data["Created"][i], dt_str).year)
year_arr = pd.Series(year_arr)
```

```python
year_arr.unique()
```

```python
yog = pd.Series(np.zeros([data.shape[0]], dtype="int64"))
```

```python
valid_arr = []
for i in range(data.shape[0]):
    i_data = {}
    i_data["pos"] = data.loc[i, "Position"]
    i_data["acayr"] = data.loc[i, "Academic Year"]
    i_data["altyr"] = data.loc[i, "What is your current academic year?"]
    valid = -1
    for j in i_data:
        if not isinstance(i_data[j], str):
            if not math.isnan(i_data[j]):
                valid = j
    valid_arr.append(valid)
    print(
        i,
        i_data,
        "\n\tYear of Graduation:",
        yog[i],
        "\n\tValidation:",
        valid_arr[i],
        sep="\t",
    )
```

```python
for i in ["Position", "Academic Year", "What is your current academic year?"]:
    print(data[i].unique())
```

```python
for i in range(len(data)):
    if valid_arr[i] == -1:
        continue
    elif valid_arr[i] == "altyr":
        match data.loc[i, "What is your current academic year?"]:
            case 'Final Year':
                yog[i] = year_arr[i]
            case '1st Year':
                yog[i] = year_arr[i] + 3
            case '2nd Year':
                yog[i] = year_arr[i]+2
            case '3rd Year':
                yog[i] = year_arr[i]+1
            case _:
                continue
    elif valid_arr[i] == "acayr":
        match data.loc[i, "Academic Year"]:
            case 4.:
                yog[i] = year_arr[i]
            case 1.:
                yog[i] = year_arr[i] + 3
            case 2.:
                yog[i] = year_arr[i]+2
            case 3.:
                yog[i] = year_arr[i]+1
            case _:
                continue
    elif valid_arr[i] == "pos":
        match data.loc[i, "Position"]:
            case 'S.Y(COMPUTER ENGG)':
                yog[i] = year_arr[i] + 2
            case 'S.Y(COMPUTER ENGINEERING)':
                yog[i] = year_arr[i] + 2
            case 'T. Y. ( C. E.)':
                yog[i] = year_arr[i] + 1
            case'T.Y. Computer Science':
                yog[i] = year_arr[i] + 1
            case 'compter Engineering Third year':
                yog[i] = year_arr[i] + 1
            case _:
                continue
```

```python
yog.replace({0: pd.NA}, inplace=True)
```
